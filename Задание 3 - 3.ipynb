```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize
```

```python
#  f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

def h1(x):
    return int(f(x))

h = np.vectorize(h1)
```

```python
a = np.arange(1, 30.5, 0.5)
```

```python
plt.plot(a, h(a), 'tab:blue', label='$ f(x) $', linewidth=2)

plt.legend(loc=1)
plt.title('график')
plt.xlabel('$ x $')
plt.ylabel('$ y $');
```

```python
x1 = optimize.minimize(h, 30, method='BFGS').fun
x2 = optimize.differential_evolution(h, [(1, 30)]).fun
```

```python
result = f'{x1[0]} {x2}'
```

```python
with open('submission-3.txt', 'w') as file:
    file.write(result)
```
